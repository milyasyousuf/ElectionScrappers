```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python
driver = webdriver.Chrome('/home/mi/Desktop/temp/Projects/chromedriver')
```

```python
driver.implicitly_wait(15) # seconds
```

```python
party = []
name = []
halka = []
count = {}
```

```python
df = pd.read_csv("dataset/2018/listOfPartiesWithSymbols.csv")
partyShortForm =pd.Series(df.shortForm.values,index=df.NameofPoliticalParty).to_dict()
```

```python
for i in range(1,273):
    hal = "na-"+str(i)
    try:
            url= "http://www.merawatan.pk/elections/"+hal
            print(url)
            driver.get(url)
            tbody = driver.find_element_by_tag_name('tbody')
            data =tbody.find_elements_by_tag_name('td')
            temp = []
            for i in data:
                if i.text!="":
                    temp.append(i.text)
            for i in range(0,len(temp),3):
                halka.append(hal)
                name.append(temp[i+1])
                if temp[i+2] in partyShortForm:
                    #print(temp[i+2]+"-----"+partyShortForm[temp[i+2]])
                    party.append(partyShortForm[temp[i+2]])
                else:
                    #print(temp[i+2])
                    party.append(temp[i+2])
            print(len(temp)/3)
            count[hal] = int(len(temp)/3)
            time.sleep(5)
    except Exception as e:
        count[hal] =0
        print("empty")
            
        #break
```
