```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python
driver = webdriver.Chrome('/home/mi/Desktop/temp/Projects/chromedriver')
```

```python
driver.implicitly_wait(15) # seconds
```

```python
url= "https://www.ecp.gov.pk/ConstResult.aspx?Const_Id=NA-1&type=NA&Election_ID=10070&Election=GENERAL%20ELECTION%2025%20JUL%202018"
print(url)
driver.get(url)
```

```python
district = []
seat = []
Constituency_Title=[]
Candidate_Name = []
party = []
Votes = []
Total_Valid_Votes = []
Total_Rejected_Votes=[]
Total_Votes=[]
Total_Registered_Voters=[]
Turnout=[]
df= pd.DataFrame()
```

```python
for i in range(1,273):
    if i!=60 and i!=103:
        url= "https://www.ecp.gov.pk/ConstResult.aspx?Const_Id=NA-"+str(i)+"&type=NA&Election_ID=10070&Election=GENERAL%20ELECTION%2025%20JUL%202018"
        print(url)
        driver.get(url)
        for elem in driver.find_elements_by_xpath('.//span[@id = "ContentPlaceHolder2_lblSubHeading"]'):
            seatName =(elem.text.split(" ")[1]).replace('(','').replace(')','')
            dist = seatName.split("-")[0]
            seatId = "NA-"+str(i)
        print(dist)
        print(seatName)
        print(seatId)
        tbody = driver.find_element_by_tag_name('tbody')
        headers = tbody.find_elements_by_tag_name('tr')
        registedVotes = int(headers[0].text.encode('utf8').decode("utf-8").split("\n")[1])
        numOfVotes = int(headers[1].text.encode('utf8').decode("utf-8").split("\n")[1])
        validVotes = int(headers[2].text.encode('utf8').decode("utf-8").split("\n")[1])
        rejectedVotes = int(headers[3].text.encode('utf8').decode("utf-8").split("\n")[1])
        TurnoutNum= headers[4].text.encode('utf8').decode("utf-8").split("\n")[1]
        for row in driver.find_elements_by_css_selector("tr.tr3"):
            col = row.find_elements_by_tag_name("td")
            district.append(dist)
            seat.append(seatName)
            Constituency_Title.append(seatId)
            Candidate_Name.append(col[0].text)
            print(col[0].text)
            party.append(col[1].text)
            Votes.append(col[2].text)
            Total_Valid_Votes.append(validVotes)
            Total_Rejected_Votes.append(rejectedVotes)
            Total_Votes.append(numOfVotes)
            Total_Registered_Voters.append(registedVotes)
            Turnout.append(TurnoutNum)

        for row in driver.find_elements_by_css_selector("tr.tr1"):
            col = row.find_elements_by_tag_name("td")
            district.append(dist)
            seat.append(seatName)
            Constituency_Title.append(seatId)
            Candidate_Name.append(col[0].text)
            print(col[0].text)
            party.append(col[1].text)
            Votes.append(col[2].text)
            Total_Valid_Votes.append(0)
            Total_Rejected_Votes.append(0)
            Total_Votes.append(0)
            Total_Registered_Voters.append(0)
            Turnout.append(TurnoutNum)
        time.sleep(5)
                #break
            #break
```

```python
df['district']=district
df['Seat']= seat
df['Constituency_Title']= Constituency_Title
df['Candidate_Name']=Candidate_Name
df['Part']=party
df['Votes']=Votes
df['Total_Valid_Votes']=Total_Valid_Votes
df['Total_Rejected_Votes']= Total_Rejected_Votes
df['Total_Votes']=Total_Votes
df['Total_Registered_Voters']=Total_Registered_Voters
df['Turnout']=Turnout
```
