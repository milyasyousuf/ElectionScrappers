```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome('/home/mi/Desktop/temp/Projects/chromedriver')
driver.implicitly_wait(15) # seconds
```

```python
party = []
name = []
halka = []
count = {}
```

```python
df = pd.read_csv("dataset/2018/listOfPartiesWithSymbols.csv")
partyShortForm =pd.Series(df.shortForm.values,index=df.NameofPoliticalParty).to_dict()
```

```python
for i in range(1,91):
    hal = "na-"+str(i)
    try:
            url= "https://hamariweb.com/pakistan-election/General/2018/"+hal
            print(url)
            driver.get(url)
            soup = BeautifulSoup(driver.page_source,"html.parser")
            table = soup.find('table', id="PageContent_rpList_dlResults_0").find_all("h6")
            print(hal)
            for i in table:
                name.append(str(i).split("\n")[1])
                part =(str(i).split("\n")[3].split("\"")[1].split("(")[0])
                if part[len(part)-1:]==" ":
                    part = part[:-1]
                halka.append(hal)
                if part in partyShortForm:
                    #print(part+"-----"+partyShortForm[part])
                    party.append(partyShortForm[part])
                else:
                    print(part)
                    party.append(part)
            print(len(table))
            count[hal] = len(table)
            time.sleep(5)
    except Exception as e:
        count[hal] =0
        print("empty")
            
    #break
```

```python
print(len(name))
print(len(party))
print(len(halka))
```

```python
df = pd.DataFrame()
df['halka'] = halka
df['name'] = name
df['party'] = party
```

```python
df.to_csv("dataset/2018/hamariWebList.csv",index=False)
```

```python
value = []
key = []
for i in range(1,len(count)+1):
    value.append(count["na-"+str(i)])
    key.append("na-"+str(i))
```
